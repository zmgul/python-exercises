## DATAFRAME
* A data structure that organizes data into a 2-dimensional table of rows and columns.

```python
data = [ ["leia", 25], ["luke", 35], ["anakin", 60] ]
print(data)
print(type(data))
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame(data, columns=["name", "age"])
print(df)
```

```python
df = pd.DataFrame(data, columns=["name", "age"], index=["1","2","3"])
print(df)
```

```python
data = {
    "name": ["leia", "luke", "anakin"],
    "age": [25,35,60]
}

df = pd.DataFrame(data)
print(df)
```

## Filtering

```python
df = pd.read_csv("Data/imdb_1000.csv")
df.head()
```

```python
df.info()
```

```python
df["star_rating"]
```

```python
df["star_rating"] > 9.0
```

```python
df[df["star_rating"] > 9.0]  
```

```python
df[
    (df["star_rating"] > 8.8) & 
    (df["duration"] > 170) 
]
```

```python
df[
    (df["star_rating"] > 8.8) & 
    (df["duration"] > 170) 
][["title", "genre"]]   #wanted columns
```

### GROUP BY

```python
df.describe()
```

```python
val = df["duration"].mean()
print(val)
```

```python
df["star_rating"].count()
```

```python
df.head(10)
```

```python
# Action genre movies count
df[
    (df["genre"] ==  "Action") 
].count()
```

```python
# Action genre movies that have star rating greater than 8.0's count
df[
    (df["genre"] ==  "Action") &
    (df["star_rating"] > 8.0)
].count()
```

```python
df.groupby("genre").count() 
```

```python
df.groupby("genre")[["title","star_rating"]].count() 
```

```python
# Action genre movies that have star rating greater than 8.0's count
df[
    (df["genre"] ==  "Action") &
    (df["star_rating"] > 8.0)
].groupby("genre").count()
```
